# –ó–∞–¥–∞–Ω–∏–µ 1

–ù–∞ —Å—Ç–∞–Ω—Ü–∏—é –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è –ø–æ—Å—Ç—É–ø–∞—é—Ç –∑–∞—è–≤–∫–∏ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ü—É–∞—Å—Å–æ–Ω–∞ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º X=2 (–≤ –µ–¥–∏–Ω–∏—Ü—É –≤—Ä–µ–º–µ–Ω–∏). –ú–æ—â–Ω–æ—Å—Ç—å —Å—Ç–∞–Ω—Ü–∏–∏ –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ–±—Å–ª—É–∂–∏–≤–∞—Ç—å –Ω–µ –±–æ–ª–µ–µ 2 –∑–∞—è–≤–æ–∫ –≤ –µ–¥–∏–Ω–∏—Ü—É –≤—Ä–µ–º–µ–Ω–∏. –ù–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤ —Ç–µ—á–µ–Ω–∏–µ –¥–∞–Ω–Ω–æ–π –µ–¥–∏–Ω–∏—Ü—ã –≤—Ä–µ–º–µ–Ω–∏:
–∞) —Å—Ç–∞–Ω—Ü–∏—è –Ω–µ —Å–ø—Ä–∞–≤–∏—Ç—Å—è —Å –ø–æ—Ç–æ–∫–æ–º –∑–∞–∫–∞–∑–æ–≤ –∏ –æ–±—Ä–∞–∑—É–µ—Ç—Å—è –æ—á–µ—Ä–µ–¥—å;
–±) —Å—Ç–∞–Ω—Ü–∏—è –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è –±—É–¥–µ—Ç –ø—Ä–æ—Å—Ç–∞–∏–≤–∞—Ç—å –∏–ª–∏ —Ä–∞–±–æ—Ç–∞—Ç—å –Ω–µ –Ω–∞ –ø–æ–ª–Ω—É—é –º–æ—â–Ω–æ—Å—Ç—å; 
–≤) –Ω–∞ —Å—Ç–∞–Ω—Ü–∏–∏ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è –Ω–µ –æ–±—Ä–∞–∑—É–µ—Ç—Å—è –æ—á–µ—Ä–µ–¥—å. 


# –†–µ—à–µ–Ω–∏–µ

–ó–∞–¥–∞—á–∞ —Å–≤—è–∑–∞–Ω–∞ —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ü—É–∞—Å—Å–æ–Ω–∞ –∏ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏–µ–º –∑–∞—è–≤–æ–∫ –Ω–∞ —Å—Ç–∞–Ω—Ü–∏–∏ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è. –ù–∞–º –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è —Ç—Ä—ë—Ö —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Å–æ–±—ã—Ç–∏–π, –∏—Å—Ö–æ–¥—è –∏–∑ —Ç–æ–≥–æ, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å—Ç—É–ø–∞—é—â–∏—Ö –∑–∞—è–≤–æ–∫ –∑–∞ –µ–¥–∏–Ω–∏—Ü—É –≤—Ä–µ–º–µ–Ω–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–æ –ø–æ –∑–∞–∫–æ–Ω—É –ü—É–∞—Å—Å–æ–Ω–∞ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º Œª=2 (—ç—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞—è–≤–æ–∫ –≤ –µ–¥–∏–Ω–∏—Ü—É –≤—Ä–µ–º–µ–Ω–∏).

–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ü—É–∞—Å—Å–æ–Ω–∞:
–§–æ—Ä–º—É–ª–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ –¥–ª—è —á–∏—Å–ª–∞ —Å–æ–±—ã—Ç–∏–π k –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ –≤—Ä–µ–º–µ–Ω–∏:

$$ùëÉ(ùëã=ùëò)=\frac{ùúÜ^{ùëò}ùëí^{‚àíùúÜ}}{ùëò}$$
 
–≥–¥–µ Œª=2, –∞ k ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å—Ç—É–ø–∏–≤—à–∏—Ö –∑–∞—è–≤–æ–∫.

a) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç–∞–Ω—Ü–∏—è –Ω–µ —Å–ø—Ä–∞–≤–∏—Ç—Å—è —Å –ø–æ—Ç–æ–∫–æ–º –∑–∞–∫–∞–∑–æ–≤ –∏ –æ–±—Ä–∞–∑—É–µ—Ç—Å—è –æ—á–µ—Ä–µ–¥—å (–±–æ–ª—å—à–µ 2 –∑–∞—è–≤–æ–∫):
–≠—Ç–æ —Å–æ–±—ã—Ç–∏–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç, –µ—Å–ª–∏ –ø–æ—Å—Ç—É–ø–∞–µ—Ç –±–æ–ª–µ–µ 2 –∑–∞—è–≤–æ–∫, —Ç–æ –µ—Å—Ç—å k>2. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —ç—Ç–æ–≥–æ —Å–æ–±—ã—Ç–∏—è:

$$P(–æ—á–µ—Ä–µ–¥—å)=P(X>2)=1‚àíP(X‚â§2)$$

–±) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç–∞–Ω—Ü–∏—è –±—É–¥–µ—Ç –ø—Ä–æ—Å—Ç–∞–∏–≤–∞—Ç—å –∏–ª–∏ —Ä–∞–±–æ—Ç–∞—Ç—å –Ω–µ –Ω–∞ –ø–æ–ª–Ω—É—é –º–æ—â–Ω–æ—Å—Ç—å (–º–µ–Ω—å—à–µ 2 –∑–∞—è–≤–æ–∫):
–≠—Ç–æ —Å–æ–±—ã—Ç–∏–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç, –µ—Å–ª–∏ –ø–æ—Å—Ç—É–ø–∞–µ—Ç –º–µ–Ω—å—à–µ 2 –∑–∞—è–≤–æ–∫, —Ç–æ –µ—Å—Ç—å k<2. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —ç—Ç–æ–≥–æ —Å–æ–±—ã—Ç–∏—è:

$$P(–Ω–µ–¥–æ–∑–∞–≥—Ä—É–∑–∫–∞)=P(X<2)=P(X=0)+P(X=1)$$

–≤) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –Ω–∞ —Å—Ç–∞–Ω—Ü–∏–∏ –Ω–µ –æ–±—Ä–∞–∑—É–µ—Ç—Å—è –æ—á–µ—Ä–µ–¥—å (–Ω–µ –±–æ–ª–µ–µ 2 –∑–∞—è–≤–æ–∫):
–≠—Ç–æ —Å–æ–±—ã—Ç–∏–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç, –µ—Å–ª–∏ –ø–æ—Å—Ç—É–ø–∞–µ—Ç –Ω–µ –±–æ–ª–µ–µ 2 –∑–∞—è–≤–æ–∫, —Ç–æ –µ—Å—Ç—å k‚â§2. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —ç—Ç–æ–≥–æ —Å–æ–±—ã—Ç–∏—è:

$$P(–±–µ–∑¬†–æ—á–µ—Ä–µ–¥–∏)=P(X‚â§2)$$

# –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Ä–µ—à–µ–Ω–∏—è

1. –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ñ–æ—Ä–º—É–ª—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π.
2. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –æ—á–µ—Ä–µ–¥–∏, –º—ã –Ω–∞—Ö–æ–¥–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –∑–∞—è–≤–æ–∫ –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ 2, –∞ –∑–∞—Ç–µ–º –≤—ã—á–∏—Ç–∞–µ–º –µ—ë –∏–∑ –µ–¥–∏–Ω–∏—Ü—ã.
3. –î–ª—è –Ω–µ–¥–æ–∑–∞–≥—Ä—É–∑–∫–∏ —Å—Ç–∞–Ω—Ü–∏–∏ (–º–µ–Ω—å—à–µ 2 –∑–∞—è–≤–æ–∫) –≤—ã—á–∏—Å–ª—è–µ–º —Å—É–º–º—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –¥–ª—è 0 –∏ 1 –∑–∞—è–≤–æ–∫.
4. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Ç—Å—É—Ç—Å—Ç–≤–∏—è –æ—á–µ—Ä–µ–¥–∏ ‚Äî —ç—Ç–æ —Å—É–º–º–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –¥–ª—è 0, 1 –∏ 2 –∑–∞—è–≤–æ–∫.

```python
import math

# –ü–∞—Ä–∞–º–µ—Ç—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞
lambda_ = 2

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø–æ –∑–∞–∫–æ–Ω—É –ü—É–∞—Å—Å–æ–Ω–∞
def poisson_pmf(k, lambda_):
    return (lambda_ ** k * math.exp(-lambda_)) / math.factorial(k)

# –∞) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –æ–±—Ä–∞–∑—É–µ—Ç—Å—è –æ—á–µ—Ä–µ–¥—å (k > 2)
p_queue = 1 - (poisson_pmf(0, lambda_) + poisson_pmf(1, lambda_) + poisson_pmf(2, lambda_))

# –±) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç–∞–Ω—Ü–∏—è –±—É–¥–µ—Ç –ø—Ä–æ—Å—Ç–∞–∏–≤–∞—Ç—å –∏–ª–∏ —Ä–∞–±–æ—Ç–∞—Ç—å –Ω–µ –Ω–∞ –ø–æ–ª–Ω—É—é –º–æ—â–Ω–æ—Å—Ç—å (k < 2)
p_underload = poisson_pmf(0, lambda_) + poisson_pmf(1, lambda_)

# –≤) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –Ω–∞ —Å—Ç–∞–Ω—Ü–∏–∏ –Ω–µ –æ–±—Ä–∞–∑—É–µ—Ç—Å—è –æ—á–µ—Ä–µ–¥—å (k <= 2)
p_no_queue = poisson_pmf(0, lambda_) + poisson_pmf(1, lambda_) + poisson_pmf(2, lambda_)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print(f"–∞) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –æ–±—Ä–∞–∑—É–µ—Ç—Å—è –æ—á–µ—Ä–µ–¥—å: {p_queue:.4f}")
print(f"–±) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç–∞–Ω—Ü–∏—è –±—É–¥–µ—Ç –ø—Ä–æ—Å—Ç–∞–∏–≤–∞—Ç—å –∏–ª–∏ —Ä–∞–±–æ—Ç–∞—Ç—å –Ω–µ –Ω–∞ –ø–æ–ª–Ω—É—é –º–æ—â–Ω–æ—Å—Ç—å: {p_underload:.4f}")
print(f"–≤) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –Ω–∞ —Å—Ç–∞–Ω—Ü–∏–∏ –Ω–µ –æ–±—Ä–∞–∑—É–µ—Ç—Å—è –æ—á–µ—Ä–µ–¥—å: {p_no_queue:.4f}")
```

# –ó–∞–¥–∞–Ω–∏–µ 2

–°—Ä–æ–∫ —Å–ª—É–∂–±—ã –∏–∑–¥–µ–ª–∏—è –∏–º–µ–µ—Ç –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –æ–∂–∏–¥–∞–Ω–∏–µ–º 2500 —á–∞—Å–æ–≤. –ù–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–π —Å—Ä–æ–∫ —Å–ª—É–∂–±—ã –¥–ª—è 100 –∏–∑–¥–µ–ª–∏–π —Å–æ—Å—Ç–∞–≤–∏—Ç –Ω–µ –º–µ–Ω–µ–µ 2300 —á–∞—Å–æ–≤

# –†–µ—à–µ–Ω–∏–µ

–°—Ä–æ–∫ —Å–ª—É–∂–±—ã –∏–∑–¥–µ–ª–∏—è –∏–º–µ–µ—Ç –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –æ–∂–∏–¥–∞–Ω–∏–µ–º Œº=2500 —á–∞—Å–æ–≤.

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –¥–ª—è –≤—ã–±–æ—Ä–∫–∏ –∏–∑ 100 –∏–∑–¥–µ–ª–∏–π —Ç–∞–∫–∂–µ –±—É–¥–µ—Ç —Ä–∞–≤–Ω–æ 2500, –∞ –≤–æ—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è –≤—ã–±–æ—Ä–∫–∏ –∏–∑ 100 –∏–∑–¥–µ–ª–∏–π —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è –≤ n —Ä–∞–∑.

–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ä–æ–∫–∞ —Å–ª—É–∂–±—ã –¥–ª—è 100 –∏–∑–¥–µ–ª–∏–π –±—É–¥–µ—Ç –∏–º–µ—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏:

$$ ùúá_{—Å—Ä}=ùúá=2500$$
–∏
$$ùúé_{—Å—Ä}={\frac{ùúé}{\sqrt{ùëõ}}}={\frac{ùúá}{\sqrt{ùëõ}}}={\frac{2500}{\sqrt{100}}}=250$$
–ì–¥–µ œÉ=2500 ‚Äî —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è –æ–¥–Ω–æ–≥–æ –∏–∑–¥–µ–ª–∏—è.

–ù–∞—Å –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–π —Å—Ä–æ–∫ —Å–ª—É–∂–±—ã –¥–ª—è 100 –∏–∑–¥–µ–ª–∏–π —Å–æ—Å—Ç–∞–≤–∏—Ç –Ω–µ –º–µ–Ω–µ–µ 2300 —á–∞—Å–æ–≤. –¢–æ –µ—Å—Ç—å –Ω–∞–º –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ X‚â•2300.

–î–ª—è —ç—Ç–æ–≥–æ –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ:

$$ùëç={\frac{{\overline{X}}-ùúá_{—Å—Ä}}{ùúé_{—Å—Ä}}}$$
‚Äã–ù–∞–π–¥—ë–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. –î–ª—è —ç—Ç–æ–≥–æ:
1. –†–∞—Å—Å—á–∏—Ç–∞–µ–º Z-–æ—Ü–µ–Ω–∫—É –¥–ª—è 2300 —á–∞—Å–æ–≤.
2. –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏.

```python
import scipy.stats as stats

# –í–≤–æ–¥ –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
mu = 2500  # –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ
sigma = 250  # —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ —Å—Ä–æ–∫–∞ —Å–ª—É–∂–±—ã –¥–ª—è –≤—ã–±–æ—Ä–∫–∏ –∏–∑ 100 –∏–∑–¥–µ–ª–∏–π
x = 2300  # –∑–Ω–∞—á–µ–Ω–∏–µ, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –∏—â–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ Z-–æ—Ü–µ–Ω–∫–∏
z = (x - mu) / sigma

# –†–∞—Å—á–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ç–æ–≥–æ, —á—Ç–æ —Å—Ä–æ–∫ —Å–ª—É–∂–±—ã –±—É–¥–µ—Ç –±–æ–ª—å—à–µ 2300 (P(X >= 2300))
probability = 1 - stats.norm.cdf(z)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
probability
```
