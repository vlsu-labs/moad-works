# Задание
В ГАИ было проведена процедура экзамена на знание ПДД. Выяснилось, что его успешно сдали 25% женщин, 50% мужчин. Среди испытуемых были 60% женщин, 30% мужчин и 10% лиц, лишенных ранее права управления автомобилем. Найти вероятность того, что случайный испытуемый успешно сдаст экзамен.
Построить дерево вероятностей:

# Решение

Для решения задачи методом "дерево вероятностей" в Jupyter Notebook необходимо:

1. Построить дерево вероятностей:

Рассмотрим три группы испытуемых: женщины, мужчины и лица, лишенные права управления автомобилем.
Для каждой группы у нас есть вероятности сдать экзамен (успех) и не сдать (неудача).
Найти полные вероятности для каждого случая (успех и неудача).

2. Найти полную вероятность того, что случайный испытуемый успешно сдаст экзамен, используя формулу полной вероятности.

3. Построить дерево вероятностей с помощью графических инструментов библиотеки matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

# Вероятности
P_W = 0.6    # Вероятность того, что случайный испытуемый — женщина
P_M = 0.3    # Вероятность того, что случайный испытуемый — мужчина
P_L = 0.1    # Вероятность того, что случайный испытуемый лишен права управления

P_W_success = 0.25  # Вероятность успешной сдачи для женщин
P_M_success = 0.5   # Вероятность успешной сдачи для мужчин
P_L_success = 0     # Вероятность успешной сдачи для лиц, лишенных права управления

# Формула полной вероятности для успеха
P_success = (P_W * P_W_success) + (P_M * P_M_success) + (P_L * P_L_success)

print(f"Полная вероятность того, что случайный испытуемый успешно сдаст экзамен: {P_success:.4f}")

# Построение дерева вероятностей
fig, ax = plt.subplots(figsize=(10, 6))

# Дерево вероятностей
# Первый уровень
ax.text(0.1, 0.9, 'Все испытуемые', fontsize=14, ha='center')

# Второй уровень
ax.text(0.3, 0.7, 'Женщины\n(60%)', fontsize=12, ha='center')
ax.text(0.3, 0.5, 'Мужчины\n(30%)', fontsize=12, ha='center')
ax.text(0.3, 0.3, 'Лишенные права\n(10%)', fontsize=12, ha='center')

# Третий уровень для женщин
ax.text(0.6, 0.75, 'Успех\n(25%)', fontsize=10, ha='center')
ax.text(0.6, 0.65, 'Неудача\n(75%)', fontsize=10, ha='center')

# Третий уровень для мужчин
ax.text(0.6, 0.55, 'Успех\n(50%)', fontsize=10, ha='center')
ax.text(0.6, 0.45, 'Неудача\n(50%)', fontsize=10, ha='center')

# Третий уровень для лишенных права
ax.text(0.6, 0.35, 'Успех\n(0%)', fontsize=10, ha='center')
ax.text(0.6, 0.25, 'Неудача\n(100%)', fontsize=10, ha='center')

# Соединительные линии
ax.plot([0.1, 0.3], [0.9, 0.7], 'k-')
ax.plot([0.1, 0.3], [0.9, 0.5], 'k-')
ax.plot([0.1, 0.3], [0.9, 0.3], 'k-')

ax.plot([0.3, 0.6], [0.7, 0.75], 'k-')
ax.plot([0.3, 0.6], [0.7, 0.65], 'k-')

ax.plot([0.3, 0.6], [0.5, 0.55], 'k-')
ax.plot([0.3, 0.6], [0.5, 0.45], 'k-')

ax.plot([0.3, 0.6], [0.3, 0.35], 'k-')
ax.plot([0.3, 0.6], [0.3, 0.25], 'k-')

ax.axis('off')
plt.show()
```

# Задание
X школьников посещает М бассейнов равновероятно. Найти вероятность того, что хотя бы в одну фирму никто не обратится, если: 
1. N = 10, М = 3; 
2. N = 8, М = 3; 
3. N = 5, М = 2.

# Решение
Пусть N — количество школьников, M — количество бассейнов. Вероятность того, что хотя бы один бассейн останется пустым, можно найти следующим образом:
1.	Определим общее количество вариантов, как можно распределить N школьников по M бассейнам. Это M в степени N.
2.	Рассмотрим вероятность того, что каждый бассейн будет посещен хотя бы одним школьником. Эта задача решается с использованием включений-исключений.
Вероятность того, что хотя бы один бассейн пустой:

P(хотя бы один пустой) = 1 − P(все бассейны заполнены)

Рассмотрим каждый из случаев.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb

def probability_no_empty_pool(N, M):
    """
    Функция для вычисления вероятности того, что хотя бы один бассейн будет пустым.
    """
    # Общее количество способов распределить N школьников по M бассейнам
    total_ways = M ** N
    
    # Используем принцип включений-исключений для подсчета числа вариантов,
    # при которых все бассейны заполнены.
    filled_ways = 0
    for k in range(M):
        filled_ways += (-1) ** k * comb(M, k) * (M - k) ** N
    
    # Вероятность, что хотя бы один бассейн будет пустым
    probability = 1 - filled_ways / total_ways
    
    return probability

# Параметры для задач a, b, и c
params = [(10, 3), (8, 3), (5, 2)]

# Вычислим вероятности
probabilities = [probability_no_empty_pool(N, M) for N, M in params]

# Выведем результаты
for i, (N, M) in enumerate(params):
    print(f"Случай {chr(97 + i)}: N = {N}, M = {M}, вероятность = {probabilities[i]:.4f}")

# Визуализация вероятностей для различных N и M
N_vals = np.arange(1, 15)
M_vals = [2, 3, 4, 5]
colors = ['blue', 'green', 'orange', 'red']

plt.figure(figsize=(10, 6))
for i, M in enumerate(M_vals):
    probabilities = [probability_no_empty_pool(N, M) for N in N_vals]
    plt.plot(N_vals, probabilities, label=f'M = {M}', color=colors[i])

plt.xlabel('Количество школьников (N)')
plt.ylabel('Вероятность хотя бы одного пустого бассейна')
plt.title('Вероятность хотя бы одного пустого бассейна при различных M')
plt.legend()
plt.grid(True)
plt.show()
```
